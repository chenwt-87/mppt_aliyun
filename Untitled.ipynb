```python
from sklearn.cluster import DBSCAN
from sklearn import metrics   
# from sklearn.preprocessing import StandardScaler
his_data = pd.read_csv('data/data_for_train_A2C.csv',index_col=0)


his_data['voltage'] = his_data['voltage'] /1e3
his_data['current'] = his_data['current'] /1e3
his_data['power'] = his_data.apply(lambda x: x['voltage']*x['current'], axis=1)
X = np.array(his_data[['voltage','power']])
# X = StandardScaler().fit_transform(X) 
```

```python
X.shape
```

```python
db = DBSCAN(eps=0.3, min_samples=4,algorithm='auto',metric='minkowski',p=2).fit(X) #DBSCAN matric = ""
labels = db.labels_  #Xlabels -1

print(':')
print(labels)

raito = len(labels[labels[:] == -1]) / len(labels)  #
print(':', format(raito, '.2%'))

n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)  # 

print(': %d' % n_clusters_)
print(": %0.3f" % metrics.silhouette_score(X, labels)) #

for i in range(n_clusters_):
    # print(' ', i, ':')
    one_cluster = X[labels == i]
    # print(one_cluster)
    plt.plot(one_cluster[:,0],one_cluster[:,1],'o',alpha=0.6,markersize=0.5)
plt.xlim(0,1.2*his_data.voltage.max())
plt.show()


```

```python
one_cluster =X[labels==4]
plt.plot(one_cluster[:,0],one_cluster[:,1],'o',alpha=0.6,markersize=0.5)
plt.xlim(0,1.2*his_data.voltage.max())
plt.ylim(0,1.2*his_data.power.max())
```

```python
his_data['label']=labels
his_data=his_data[his_data.label!=-1]

his_data['voltage'] = his_data['voltage'] *1e3
his_data['current'] = his_data['current'] *1e3
his_data.reset_index(inplace=True)
his_data.to_csv('data/data_for_train_A2C_with_label.csv')
```
